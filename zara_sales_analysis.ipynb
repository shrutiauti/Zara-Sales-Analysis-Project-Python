## 5. Feature Engineering (Creating New Columns)

```python
def price_range(price):
    if price < 40 :
        return 'Low'
    elif price > 40 and price < 80:
        return 'Medium'
    else:
        return 'High'

zara['Price_Range']= zara['price'].apply(price_range)
zara[['price','Price_Range']]
```

```python
zara['Promotion_Flag'] = zara['Promotion'].map({'Yes':1,'NO':0})

zara[['Promotion','Promotion_Flag']]
```

```python
zara["Total"] = zara["price"] * zara["Sales Volume"]
zara.head()
```

### ii. Qualitative or Categorical Variable 

#### In this step, we will explore the relationships between the variables, particularly focusing on qualitative variables. We'll see the relationships between:
* Promotion and Seasonal
* Promotion and terms
* Promotion and section
* Promotion and Product Position
* Product Position and section
* Position and terms

```python
from scipy.stats import chi2_contingency
import math
import numpy as np
```

```python
observed = pd.crosstab(zara['Promotion'], zara['Seasonal'])
chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value : 1503.22
* p value : 0.0

* chi square test conclude that there is relationship between promotion and seasonal . This means that whether a product is on promotion depends on whether it is seasonal or not. Seasonal products are more likely to be promoted than non-seasonal products

```python
observed= pd.crosstab(zara['Promotion'],zara['terms'])

chi2, p, dof, expected=chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value :  4.148
* p value :  0.39

* chi square test conclude that there is no relationship between promotion and terms. In other words, promotions are independent of the product terms , whether a product is promoted does not depend on its terms.

```python
observed= pd.crosstab(zara['Promotion'],zara['section'])

chi2, p, dof, expected=chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value :  0.8
* p value : 0.37

* chi square test conclude that there is no relationship between promotion and section.The product that is on promotion can belong to any section. This means a promotion can happen in any section; it is not limited to a particular section.

```python
observed= pd.crosstab(zara['Promotion'],zara['Product Position'])

chi2, p, dof, expected=chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value : 9.14
* p value : 0.01

* chi square test conclude that there is relationship between promotion and Product Position. This means whether a product is on promotion depends on its position some positions are more likely to have promotions than others. If a promoted product is placed in a visible or front position, it is more likely to attract customers and increase sales.

```python
observed= pd.crosstab(zara['section'],zara['Product Position'])

chi2, p, dof, expected=chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value : 1.48
* p value : 0.48

* chi square test conclude that there is no relationship between section and Product Position. This means a product’s position (Aisle, End-cap, Front of Store) does not depend on the section it belongs to.  

```python
observed= pd.crosstab(zara['terms'],zara['Product Position'])

chi2, p, dof, expected=chi2_contingency(observed)

print("Chi-squared test results:")
print("Chi-squared value:", round(chi2,2))
print("P-value:", round(p,2))
print("Degrees of freedom:",  round(dof,2))
print("Expected frequencies:\n", expected )
```

### Conclusion 
* chi square value : 8.37
